```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_labels():
    X_names = []
    y = []
    depths = [{},{}]
    with open('./dataset/Outputs/labels.txt') as f:
        for line in f:
            infos = line.split(' ')
            name, infos = infos[0], infos[1]
            infos = [float(x) for x in infos.split(',')]
            if infos[1] != 220.0:
                if infos[0] not in depths[0]:
                    depths[0][infos[0]] = 0
                else:
                    depths[0][infos[0]] += 1
                if infos[1] not in depths[1]:
                    depths[1][infos[1]] = 0
                else:
                    depths[1][infos[1]] += 1

                y.append(infos)
                X_names.append(name)
    return X_names, y, depths
```

```python
X, y, depths= load_labels()
y = np.array(y)
print(len(X), y.shape)
print(depths)
```

```python
# A function to decress the number of labels
def decrease_label(y):
    
    for line in y:
        
        if line[0] <= -120:
            line[0] = 1
        elif line[0] <= 80:
            line[0] = 2
        elif line[0] <= 200:
            line[0] = 3
            
        if line[1] <= -100:
            line[1] = 1
        elif line[1] <= 60.0:
            line[1] = 2
        elif line[1] <= 180.0:
            line[1] = 3
        
    return y
```

```python
y = decrease_label(y)
```

```python
def preprocess_offset_data(y):
    
    y_ = y[:, 0:2]+200    
    enc = OneHotEncoder()
    y__ = enc.fit_transform(y_).toarray()
    
    return np.array(y__)
    
y_offset = preprocess_offset_data(y)   
print(y_offset.shape)
```

```python
def get_batch(X_train, y_train, i, batch_size=64, img_size=(128, 128)):
    
    if i*batch_size+batch_size <= len(X_train):
        X_batch_names = X_train[i*batch_size:i*batch_size+batch_size]
        y_batch = y_train[i*batch_size: i*batch_size+batch_size]
    else:
        X_batch_names = X_train[i*batch_size:len(X_train)] # last block
        y_batch = y_train[i*batch_size: len(y_train)]
    
    X_batch = []
    i = 0
    for name in X_batch_names:
        try:
            img = Image.open('./dataset/Outputs/'+name)
            if img.mode != 'RGB':
                img = img.convert('RGB')
                
            img = np.array(img.resize(img_size))
            X_batch.append(img)
            i+=1
            if i%10000==0:
                print('Done with '+str(i)+' images')
        except Exception as e:
            print(e)
            continue
    
    return np.array(X_batch), np.array(y_batch[:, 0:3]), np.array(y_batch[:, 3:6])
```

```python
batch_size = X_train.shape[0]
num_of_batches = len(X_train)/batch_size
img_size = (128, 256)
print('# of batches: '+str(num_of_batches))
X_batch, y_batch_0, y_batch_1 = get_batch(X_train, y_train, 0, batch_size=batch_size, img_size=img_size)
```

```python
def get_test_batch(X_test, y_test, img_size=(128, 128)):
    
    X_batch = []
    y_batch_1 = []
    y_batch_2 = []
    for i in range(len(X_test)):
        name = X_test[i]
        try:
            img = Image.open('./dataset/Outputs/'+name)
            if img.mode != 'RGB':
                img = img.convert('RGB')
                
            img = np.array(img.resize(img_size))
            X_batch.append(img)
            y_batch_1.append(y_test[i, 0:3])
            y_batch_2.append(y_test[i, 3:6])
        except Exception as e:
            print(e)
            continue
            
    return X_batch, y_batch_1, y_batch_2
```

```python
def get_acc(results, y_test_batch_1, y_test_batch_2):
    count = 0.0
    count_1 = 0.0
    count_2 = 0.0
    for i in range(y_test_batch_1.shape[0]):
        if results[i]==(y_test_batch_1[i], y_test_batch_2[i]):
            count += 1
        if results[i][0]==y_test_batch_1[i]:
            count_1 += 1
        if results[i][1]==y_test_batch_2[i]:
            count_2 += 1
        print(results[i], (y_test_batch_1[i], y_test_batch_2[i]), results[i]==(y_test_batch_1[i], y_test_batch_2[i]))
    return count/y_test_batch_1.shape[0], count_1/y_test_batch_1.shape[0], count_2/y_test_batch_1.shape[0]
test_acc = get_acc(results, y_test_batch_1, y_test_batch_2)
print(test_acc)
```
