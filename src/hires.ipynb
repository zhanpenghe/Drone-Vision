```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from PIL import Image
```

```python
import os
def load_dataset(path, img_size):
    
    file_list = os.listdir(path)
    labels = []
    images = []
    
    for filename in file_list:
        # print(filename)
        try:
            label = filename.split('_')[0]
            img = Image.open(path+filename)
            
            img = np.array(img.resize(img_size))

            if label == 'Left':
                labels.append([1])
                images.append(img)
            elif label == 'Middle':
                labels.append([2])
                images.append(img)
            elif label == 'Right':
                labels.append([3])
                images.append(img)
            
        except Exception as e:
            print(e)
            continue
    return np.array(images), np.array(labels)
```

```python
img_shape = (128, 128)

X, y = load_dataset('./dataset/hires_output/Outputs/', img_size=img_shape)

print(X.shape, y.shape)
print(y[0])
print(y[1000])

X_temp_img = Image.fromarray(X[5])
X_temp_img
```

```python
enc = OneHotEncoder()
y = enc.fit_transform(y).toarray()
print(y[0])
print(y[1000])
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.03, random_state=42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from resNet.build_network import build_network_for_hires
```

```python
model = build_network_for_hires((128, 128, 3), 3)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())
```

```python
model.load_weights('hires_w1.h5')
```

```python
#model.fit(x=X_train, y=y_train, epochs=30, validation_data=[X_test, y_test], batch_size=512)
```

```python
def get_acc(y, predicted_y):
    result = []
    count = 0
    for i in range(y.shape[0]):
        result.append([np.argmax(y[i]), np.argmax(predicted_y[i]), np.argmax(y[i])==np.argmax(predicted_y[i])])
        if np.argmax(y[i])==np.argmax(predicted_y[i]):
            count+=1
    return result, count, count/y.shape[0]
```

```python
training_predicted = model.predict(X_train, batch_size=512)
re, ct, acc = get_acc(y_train, training_predicted)
print('Training accuracy:\nTotal # of samples: %d\nTotal # of correct predictions: %d\nAccuracy: %f\n'%(y_train.shape[0], ct, acc))

test_predicted = model.predict(X_test)
re, ct, acc = get_acc(y_test, test_predicted)
print('Training accuracy:\nTotal # of samples: %d\nTotal # of correct predictions: %d\nAccuracy: %f\n'%(y_test.shape[0], ct, acc))
```
